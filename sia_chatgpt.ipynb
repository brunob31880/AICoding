```python
import json
import xml.etree.ElementTree as ET
```

```python
# Ouverture du fichier XML
tree = ET.parse('datas/XML_SIA_2022-12-29.xml')
root = tree.getroot()
```

```python
# Initialisation du dictionnaire vide
parties=[]
# Parcours des éléments du fichier XML
for partie in root.iter('Partie'):
     # Récupération des attributs pk et lk
    pk = partie.get('pk')
    lk = partie.get('lk')
    # Récupération de la valeur de l'élément NomUsuel s'il existe
    nom_usuel = partie.find('NomUsuel')  
    if nom_usuel is not None:
        nom_usuel = nom_usuel.text
    # Récupération de la valeur de l'élément NumeroPartie s'il existe
    numero_partie = partie.find('NumeroPartie')
    if numero_partie is not None:
        numero_partie = numero_partie.text
    # Récupération de la valeur de l'élément NomPartie s'il existe
    nom_partie = partie.find('NomPartie')
    if nom_partie is not None:
        nom_partie = nom_partie.text    
    # Récupération de la valeur de l'élément Espace s'il existe
    espace = partie.find('Espace')
    if espace is not None:
        espace_pk = espace.get('pk')
        espace_lk = espace.get('lk')
    else:
        espace_pk = None
        espace_lk = None      
    coords = []    
    # Récupération de la valeur de l'élément Geometrie s'il existe
    geometrie=partie.find('Geometrie')
    if geometrie is not None:
        for ligne in geometrie.text.split('\n'):
            lat, lon = ligne.split(',')
            coords.append((float(lat), float(lon)))   
    # Création de l'objet Python
    objet_partie = {
        'pk': pk,
        'lk': lk,
        'espace_pk': espace_pk,
        'espace_lk': espace_lk,
        'nom_partie': nom_partie,
        'numero_partie': numero_partie,
        'nom_usuel': nom_usuel,
        'geometrie': coords
    }
    #print(objet_partie)
    # Ajout de l'objet partie au dictionnaire avec la clé lk
    parties.append(objet_partie)

# Enregistrement de la liste dans le fichier JSON
with open('parties.json', 'w') as f:
    json.dump(parties, f)
```

```python
# Parcourir les éléments <Espace> et créer des dictionnaires Python
espaces = []
for espace_elem in root.iter('Espace'):
    espace_dict = {}
    espace_dict['pk'] = int(espace_elem.get('pk'))
    espace_dict['lk'] = espace_elem.get('lk')
    territoire_elem = espace_elem.find('Territoire')
    if territoire_elem is not None:
        espace_dict['territoire_pk'] = int(territoire_elem.get('pk'))
        espace_dict['territoire_lk'] = territoire_elem.get('lk')
    type_espace_elem = espace_elem.find('TypeEspace')
    if type_espace_elem is not None:
        espace_dict['type_espace'] = type_espace_elem.text
    nom_elem = espace_elem.find('Nom')
    if nom_elem is not None:
        espace_dict['nom'] = nom_elem.text
    altr_ft_elem = espace_elem.find('AltrFt')
    if altr_ft_elem is not None:
        espace_dict['altr_ft'] = int(altr_ft_elem.text)
    ad_associe_elem = espace_elem.find('AdAssocie')
    if ad_associe_elem is not None:
        espace_dict['ad_associe_pk'] = int(ad_associe_elem.get('pk'))
        espace_dict['ad_associe_lk'] = ad_associe_elem.get('lk')
    espaces.append(espace_dict)
with open('espaces.json', 'w') as f:
    json.dump(espaces, f)
```

```python
routes = []
for route in root.iter('Route'):
    pk = route.get('pk')
    lk = route.get('lk')
    territoire = route.find('Territoire')
    if territoire is not None:
        territoire_pk = territoire.get('pk')
        territoire_lk = territoire.get('lk')
    else:
        territoire_pk = None
        territoire_lk = None
    prefixe = route.find('Prefixe').text if route.find('Prefixe') is not None else None
    numero = route.find('Numero').text if route.find('Numero') is not None else None
    origine = route.find('Origine')
    if origine is not None:
        origine_pk = origine.get('pk')
        origine_lk = origine.get('lk')
    else:
        origine_pk = None
        origine_lk = None
    route_type = route.find('RouteType').text if route.find('RouteType') is not None else None
    type_compte_rendu = route.find('TypeCompteRendu').text if route.find('TypeCompteRendu') is not None else None
    
    route_dict = {
        'pk': pk,
        'lk': lk,
        'Territoire_pk': territoire_pk,
        'Territoire_lk': territoire_lk,
        'Prefixe': prefixe,
        'Numero': numero,
        'Origine_pk': origine_pk,
        'Origine_lk': origine_lk,
        'RouteType': route_type,
        'TypeCompteRendu': type_compte_rendu
    }
    routes.append(route_dict)

with open('routes.json', 'w') as f:
    json.dump(routes, f)
```

```python
services=[]
# Parcours de toutes les balises <Service>
for service in root.iter('Service'):
    # Création d'un dictionnaire pour stocker les attributs
    service_dict = {}
    # Récupération de l'attribut 'pk' s'il existe
    pk = service.get('pk')
    if pk:
        service_dict['pk'] = pk
    # Récupération de l'attribut 'lk' s'il existe
    lk = service.get('lk')
    if lk:
        service_dict['lk'] = lk
    # Récupération des autres attributs s'ils existent
    indic_lieu = service.find('IndicLieu')
    if indic_lieu is not None:
        service_dict['IndicLieu'] = indic_lieu.text
    indic_service = service.find('IndicService')
    if indic_service is not None:
        service_dict['IndicService'] = indic_service.text
    langue = service.find('Langue')
    if langue is not None:
        service_dict['Langue'] = langue.text
    services.append(service_dict)
with open('services.json', 'w') as f:
    json.dump(services, f)
```

```python
# Liste de dictionnaires pour stocker les données des éléments <Volume>
volumes = []
# Boucle sur les éléments <Volume>
for volume in root.findall(".//Volume"):
    # Création d'un dictionnaire pour stocker les données de l'élément <Volume>
    volume_dict = {}
    # Ajout des attributs du noeud <Volume> dans le dictionnaire
    volume_dict["pk"] = volume.get("pk")
    volume_dict["lk"] = volume.get("lk")

    # Récupération de l'élément <Partie> s'il existe
    partie = volume.find("Partie")
    if partie is not None:
        volume_dict["partie_pk"] = partie.get("pk")
        volume_dict["partie_lk"] = partie.get("lk")

    # Récupération des autres éléments s'ils existent
    sequence = volume.find("Sequence")
    if sequence is not None:
        volume_dict["sequence"] = sequence.text

    plafond_ref_unite = volume.find("PlafondRefUnite")
    if plafond_ref_unite is not None:
        volume_dict["plafond_ref_unite"] = plafond_ref_unite.text

    plafond = volume.find("Plafond")
    if plafond is not None:
        volume_dict["plafond"] = plafond.text

    plancher_ref_unite = volume.find("PlancherRefUnite")
    if plancher_ref_unite is not None:
        volume_dict["plancher_ref_unite"] = plancher_ref_unite.text

    plancher = volume.find("Plancher")
    if plancher is not None:
        volume_dict["plancher"] = plancher.text

    hor_code = volume.find("HorCode")
    if hor_code is not None:
        volume_dict["hor_code"] = hor_code.text

    hor_txt = volume.find("HorTxt")
    if hor_txt is not None:
        volume_dict["hor_txt"] = hor_txt.text

    activite = volume.find("Activite")
    if activite is not None:
        volume_dict["activite"] = activite.text

    remarque = volume.find("Remarque")
    if remarque is not None:
        volume_dict["remarque"] = remarque.text
    # Ajout du dictionnaire dans la liste de dictionnaires
    volumes.append(volume_dict)
with open('volumes.json', 'w') as f:
    json.dump(volumes, f)
```
